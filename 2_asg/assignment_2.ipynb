```python
"""
Programming exercise 2: Pandas Data frames
Here you can find practical task for tutorial session 2. 

Please note, that the feedback and discussion on this are only are happening at tutorials. 

Creating Data Frame
You are given a fragment of Data about medals from the Olympics
 
 
Country:    Gold    Silver  Bronze
NOR         16      8       13
GER         12      10      5
CHN         9       4       2
 

1. Create a pandas Data frame from this data
2. Select the first and third rows and the first and third columns
3. Return a data frame that contains just the CHN row
4. Return a data frame that contains the medals of Norway and Germany.
5. *Select countries who won ten or more gold medals and return the Country and Gold columns*
6. Sort the medals by bronze (descending)
"""
import pandas as pd
data = {"Country": ["NOR", "GER", "CHN"], "Gold": [16, 12, 9], "Silver": [8, 10, 4], "Bronze": [13, 5, 2]}
medals = pd.DataFrame(data)
# print(data)
# print(medals)
obj = medals.iloc[[0,2],[0,2]]
print(obj)
print(medals["Country"]=="CHN")
print(medals[medals["Country"]=="CHN"])
print(medals[medals["Country"].isin(["GER", "NOR"])])
medals[medals["Gold"]>10][["Country","Gold"]]
sorted_by_bronze = medals.sort_values(by="Bronze", ascending=False)
sorted_by_bronze

```
