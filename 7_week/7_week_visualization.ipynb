```python
import pandas as pd
df = pd.read_csv("result.csv")
```

```python
df.head(5)
```

```python
import matplotlib.pyplot as plt

# Assuming your dataframe is named df and it contains 'GDP per capita' and 'Happiness score'
plt.figure(figsize=(10, 6))
plt.scatter(df['GDP per capita'], df['Score'], color='blue', alpha=0.5)

# Adding labels and title
plt.title('Happiness Score vs GDP per capita')
plt.xlabel('GDP per capita')
plt.ylabel('Happiness score')

# Show the plot
plt.show()
```

```python
continent_avg = df.groupby("continent")["Score"].mean()
```

```python
continent_avg
```

```python

# Assuming your dataframe is named df and it contains 'Happiness score' and 'Continent'

# Group by 'Continent' and calculate the average 'Happiness score' per continent
continent_avg = df.groupby('continent')['Score'].mean()

# Sort by average happiness score if you want to display it in descending order
continent_avg = continent_avg.sort_values(ascending=False)

# Plotting the bar chart
plt.figure(figsize=(10, 6))
continent_avg.plot(kind='bar', color='skyblue')

# Adding labels and title
plt.title('Average Happiness Score by Continent')
plt.xlabel('Continent')
plt.ylabel('Average Happiness Score')

# Display the plot
plt.show()
```

```python
# Take Top 10 happiest countries and 10 least happy Countries - create a box plot visualization level of social support in those countries 

sorted_df = df.sort_values(by="Score", ascending=False)
```

```python
sorted_df
```

```python
top10_most_happy = sorted_df.head(10)
top10_least_happy = sorted_df.tail(10)
```

```python
# Combine both top and bottom 10 into a single dataframe with an additional column for label
top10_most_happy['Happiness Category'] = 'Top 10 Happiest'
top10_least_happy['Happiness Category'] = 'Bottom 10 Least Happy'

# Concatenate both dataframes
combined_df = pd.concat([top10_most_happy, top10_least_happy])
```

```python
combined_df
```

```python

import seaborn as sns
plt.figure(figsize=(20, 10))
sns.boxplot(data=combined_df, x='Happiness Category', y='Social support', palette='Set2')

# Add title and labels
plt.title('Social Support Level in Top 10 Happiest and Bottom 10 Least Happy Countries')
plt.xlabel('Happiness Category')
plt.ylabel('Social Support')

# Display the plot
plt.show()
```

```python

import seaborn as sns
plt.figure(figsize=(20, 10))
sns.boxplot(data=combined_df, x='Happiness Category', y='Healthy life expectancy', palette='Set2')

# Add title and labels
plt.title('Social Support Level in Top 10 Happiest and Bottom 10 Least Happy Countries')
plt.xlabel('Happiness Category')
plt.ylabel('Healthy life expectancy')

# Display the plot
plt.show()
```
